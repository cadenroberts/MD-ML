```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The path to the model to plot:
result_direcotry = "/media/DATA_18_TB_1/daniel_s/cgschnet/cgschnet_models/cg_single_chain_CAprior_2024.04.03"
```

```python
history = np.load(f'{result_direcotry}/history.npy', allow_pickle=True).item()
train_loss_list = np.array(history['train'])/history['train'][0]
val_loss_list = np.array(history['val'])/history['val'][0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Percent Original Loss')

ax1.plot(train_loss_list, label=f"training loss")
ax1.plot(val_loss_list, label=f"validation loss")

ax1.set_xlabel("Epoch")
ax1.set_ylabel("% Original MSE Loss")

ax1.set_title(f"Training & Validation Loss")

max_epoch = 0
delta_train_loss_list = train_loss_list[1:] - train_loss_list[:-1]
delta_val_loss_list = val_loss_list[1:] - val_loss_list[:-1]
max_epoch = max(max_epoch, len(delta_val_loss_list))

ax2.plot(range(1,len(delta_train_loss_list)+1), delta_train_loss_list, label=f"training loss")
ax2.plot(range(1,len(delta_train_loss_list)+1), delta_val_loss_list, label=f"validation loss")
ax2.axhline(y=0.0, color='black', linestyle='--', linewidth='0.5')

ax2.set_title(f"Change in Loss Per Epoch")
ax2.set_xlabel("Epoch")
ax2.set_ylabel(f"∆ % Loss")

ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.tight_layout()
```
